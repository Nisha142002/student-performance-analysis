```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()

```

```python
df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="gender", y="math score", data=df)
plt.title("Average Math Score by Gender")
plt.show()
```

```python
corr = df[["math score", "reading score", "writing score"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Subject Score Correlation Heatmap")
plt.show()
```

## Insights
- Female students generally score higher in reading and writing.
- Math, reading, and writing scores are strongly correlated.
- There’s a clear performance gap based on test preparation completion.
